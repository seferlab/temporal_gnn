```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np

data = np.random.rand(3, 3)

G = nx.Graph()
num_series = data.shape[1]

for i in range(num_series):
    for j in range(data.shape[0]):
        G.add_node(i * data.shape[0] + j)

for node in G.nodes:
    for node2 in G.nodes:
        if node != node2:
            G.add_edge(node, node2)
```

```python
larger_data = np.random.rand(30, 3)
larger_data[:, 0] = np.random.rand(30) * 1 + 5
larger_data[:, 1] = np.random.rand(30) * 1 + 7
larger_data[:, 2] = np.random.rand(30) * 1 + 9

plt.plot(larger_data[:, 0], color="skyblue")
plt.plot(larger_data[:, 1], color="lightgreen")
plt.plot(larger_data[:, 2], color="salmon")

# Colors
colors = ["skyblue", "lightgreen", "salmon"]

# Plot lines and mark points from index 90 to 93
for i in range(3):
    plt.plot(larger_data[:, i], color=colors[i])
    plt.scatter(
        np.arange(20, 23),  # Indices 90 to 93
        larger_data[20:23, i],  # Corresponding values
        color=colors[i],  # Fill color
        edgecolor="black",  # Optional: black border
        zorder=5,  # On top
    )

plt.ylim(3.9, 10.5)

# Add vertical dashed lines to indicate window
plt.axvline(x=19.5, color="gray", linestyle="--")
plt.axvline(x=22.5, color="gray", linestyle="--")
plt.axhline(y=6.5, color="black", linestyle="-")
plt.axhline(y=8.5, color="black", linestyle="-")
plt.axhline(y=4.5, color="black", linestyle="-")
plt.annotate(
    "",
    xy=(30, 8.5),
    xytext=(29, 8.35),
    va="center",
    ha="right",
    fontsize=10,
    color="black",
)
plt.annotate(
    "",
    xy=(30, 8.5),
    xytext=(28, 8.5),
    arrowprops=dict(arrowstyle="->", color="black", lw=1),
    va="center",
    ha="right",
    fontsize=10,
    color="black",
)
plt.annotate(
    "",
    xy=(30, 6.5),
    xytext=(29, 6.35),
    va="center",
    ha="right",
    fontsize=10,
    color="black",
)
plt.annotate(
    "",
    xy=(30, 6.5),
    xytext=(28, 6.5),
    arrowprops=dict(arrowstyle="->", color="black", lw=1),
    va="center",
    ha="right",
    fontsize=10,
    color="black",
)
plt.annotate(
    "",
    xy=(30, 4.5),
    xytext=(29, 4.35),
    va="center",
    ha="right",
    fontsize=10,
    color="black",
)
plt.annotate(
    "",
    xy=(30, 4.5),
    xytext=(28, 4.5),
    arrowprops=dict(arrowstyle="->", color="black", lw=1),
    va="center",
    ha="right",
    fontsize=10,
    color="black",
)

plt.yticks([])
plt.xticks([])

for spine in plt.gca().spines.values():
    spine.set_visible(False)

plt.show()
```

```python
pos = nx.spring_layout(G, seed=42)
node_colors = (
    ["skyblue"] * data.shape[0]
    + ["lightgreen"] * data.shape[0]
    + ["salmon"] * data.shape[0]
)

plt.figure(figsize=(8, 6))
nx.draw(G, pos, node_color=node_colors, node_size=800, font_size=10, edge_color="gray")
labels = nx.get_edge_attributes(G, "weight")
nx.draw_networkx_edge_labels(
    G, pos, edge_labels={k: f"{v:.2f}" for k, v in labels.items()}, font_size=8
)
plt.show()
```
